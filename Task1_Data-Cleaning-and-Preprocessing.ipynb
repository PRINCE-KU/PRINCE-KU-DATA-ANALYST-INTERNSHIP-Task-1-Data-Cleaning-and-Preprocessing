<H6> Step 1: Import Libraries & Load Dataset</H6>

```python
import pandas as pd
df = pd.read_csv("C:/Users/PRINCE KUMAR/OneDrive/Desktop/Customer Personality Analysis1.csv")
df.head()
```

<H6> Step 2: Check for Missing Values</H6>

```python

df.isnull().sum()
```

<H6>Step 3: Fill Missing Values in 'Income' Column</H6>

```python

median_income = df['Income'].median()
df['Income'] = df['Income'].fillna(median_income)
df.isnull().sum()

```

```python
df.head()
```

<H6> Step 4: Remove Duplicate Rows</H6>

```python

df = df.drop_duplicates()
df.head()
```

<H6>Step 5: Rename Columns (lowercase, no spaces)</H6>

```python
# Standardize column names 
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.head()
```

<H6>Step 6: Convert 'dt_customer' to Datetime Format</H6>

```python

# Convert to datetime first (if not already)
df['dt_customer'] = pd.to_datetime(df['dt_customer'], errors='coerce', dayfirst=True)

# Format date as 'dd-mm-yyyy' string format
df['dt_customer'] = df['dt_customer'].dt.strftime('%d-%m-%Y')
df.head()
```

<H6>Step 7: Clean Text Columns</H6>

```python
# Standardize values in 'education' and 'marital_status'
df['education'] = df['education'].str.strip().str.lower()
df['marital_status'] = df['marital_status'].str.strip().str.lower()
df.head()
```
